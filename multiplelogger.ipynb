```python
import logging

#create a logger for module1
logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

#create a moule for logger2

logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)



#configuring the logging
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(level)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    
    )
```

```python
#log message with different logger
logger1.debug("this is a debug message for module1")
logger2.warning("this is a warning message for module2")
logger2.error("this is error message")
```
